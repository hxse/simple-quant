```python
# get_data
import pandas as pd
from pathlib import Path

config = {
    # "name": "SHFE.rb2405",
    # "name": "CZCE.MA405",
    # "name": "DCE.m2405",
    # "name": "CZCE.RM405",
    "name": "DCE.c2405",
    # "name": "DCE.l2405",
    # "name": "CZCE.TA405",
    "frequency": "300s",  # 15s, 30s, 60s, 300s, 900s, 1800s
    "count": 6000,
    "refresh": True,
    # "end_time": None,
    "end_time": "2024-03-22 23:00:00",
}

path = Path(
    f"../../csv/{config['name']}/{config['name']} {config['frequency']} {config['count']} {config['end_time'].replace(':','_')}.csv"
)
path.parent.mkdir(parents=True, exist_ok=True)

data = get_or_read_data(config,path)
data
```

```python
# add_ta
from add_ta import add_ta

ma_list = [4, 8, 16, 32, 64, 128, 256, 512]
ma_name = "sma"
data = add_ta(data, ma_name=ma_name, ma_list=ma_list, atr_length=14)

# data_csv = data.transpose().to_csv(header=False, index=False)
data_json = data.transpose().to_json(orient="values")
data_label = data.columns.to_list()
data
```

```python
# signals
import pandas_ta as ta

ma_name = "sma"  # "ema"

GC = ta.tsignals(
    (data["close"] > data[f"{ma_name}_4"])
    & (data[f"{ma_name}_4"] < data[f"{ma_name}_64"])
    # & (data[f"{ma_name}_4"] < data[f"{ma_name}_16"])
    & (data[f"{ma_name}_4"] > data[f"{ma_name}_128"]),
    asbool=True,
)

DC = ta.tsignals(
    (data["close"] < data["sma_4"])
    & (data[f"{ma_name}_4"] > data[f"{ma_name}_64"])
    # & (data[f"{ma_name}_4"] > data[f"{ma_name}_16"])
    & (data[f"{ma_name}_4"] < data[f"{ma_name}_128"]),
    asbool=True,
)

ohlcvDataList = [0, 4, 7 + ma_list.index(128)]

GC
```

```python
# backtest
from backtest import backtest, analysis_data

# mode: sl, tsl, tsl2, tsl_atr

# backtest(GC, data, way="up", mode="sl", wait=20, stop=10)
# backtest(DC, data, way="down", mode="sl", wait=20, stop=10)

# backtest(GC, data, way="up", mode="tsl", stop=10)
# backtest(DC, data, way="down", mode="tsl", stop=10)

# backtest(GC, data, way="up", mode="tsl_atr", rate=2)
# backtest(DC, data, way="down", mode="tsl_atr", rate=2)

backtest(GC, data, way="up", mode="tsl2", wait=20, stop=10)
backtest(DC, data, way="down", mode="tsl2", wait=20, stop=10)

GC_json = GC.transpose().to_json(orient="values")
DC_json = DC.transpose().to_json(orient="values")

storeDataUp = analysis_data(data, GC[GC.columns[-2]], GC[GC.columns[-1]], mode="up")
storeDataDown = analysis_data(data, DC[DC.columns[-2]], DC[DC.columns[-1]], mode="down")
storeData = storeDataUp.join(storeDataDown, lsuffix="_up", rsuffix="_down")
storeData.drop(["time_down"],axis=1,inplace=True)
storeData.rename(columns={"time_up": "time"}, inplace=True)
storeData["money"] = storeData["money_up"] + storeData["money_down"]
storeDataJson = storeData.transpose().to_json(orient="values")
storeData
```

```python
# anywidget
from anywidget_demo import Counter

c = Counter()
c.value = 0  # 暂时没用
c.ohlcvData = data_json
c.ohlcvDataList = ohlcvDataList  # 选择显示哪列数据, 注释后隐藏图标, 4是收盘价
c.ohlcvDataLabel = data_label
c.optObj = {
    "id": "ohlcvData",
    "width": "50%",
    "height": "40%",
    "scales": {"x": {"time": True, "distr": 2}},
}

c.storeData = storeDataJson
c.storeList = [0, 1, 2, 5, 6, 9]  # 选择显示哪列数据, 注释后隐藏图标
c.storeOptObj = {
    "id": "storeData",
    "width": "50%",
    "height": "22%",
    "scales": {"x": {"time": True, "distr": 2}},
}
# c.storeData = "0,1,2,3,4\n1,2,3,4,5\n2,3,4,5,6,\n"

c.GCDCList = [4, 5]  # [2,3] [4,5] # 2,3是自动金叉死叉进出场, 4,5是止损离场
c.GCData = GC_json
c.DCData = DC_json
```
